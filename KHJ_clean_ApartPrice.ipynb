```python
import pandas as pd
from pandas import DataFrame
```

```python
xlsx1= pd.ExcelFile('data/monthlyTradingPriceIndex_apart.xlsx')
```

```python
df= xlsx1.parse(xlsx1.sheet_names[0], skiprows=10)
```

```python
df= df.iloc[1:,1:]
```

```python
f= lambda x: x.replace('년 ','-').replace('월','')
```

```python
df.columns= df.columns.map(f)
# df.columns[0]
```
