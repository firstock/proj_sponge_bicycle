### train, test 나눔

```python
brCol= bikeRent.columns
X= bikeRent.loc[:, brCol != 'rentcnt']
y= bikeRent['rentcnt']
```

```python
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test=\
    train_test_split(X, y, random_state=33)#,stratify= y)
```

```python
print('train\t',X_train.shape, y_train.shape)
print('test\t',X_test.shape, y_test.shape)
```

### 숲: 기본값

```python
bikeForest= RandomForestRegressor(n_estimators=5, random_state= 2)
```

```python
bikeForest.fit(X_train, y_train)
```

```python
print('훈련%15.5f'%bikeForest.score(X_train, y_train))
print('새거%15.5f'%bikeForest.score(X_test, y_test))
```

### 숲: 돌려보자

```python
def howMaxNesti(rangeMin= 100, rangeMax= 200, stride= 10):
    testMax= 0
    nRange= (rangeMin,rangeMax,stride)
    for n in nRange:
        bikeForest= RandomForestRegressor(n_estimators= n, random_state= 2)
        bikeForest.fit(X_train, y_train)
        forestScore= bikeForest.score(X_test, y_test) 
        if testMax< forestScore:
            testMax= forestScore
            nEstiMax= n
            trainMax= bikeForest.score(X_train, y_train)
    print('새거%15.12f'%testMax)
    print('n%15d'%nEstiMax)
    print('훈련%15.12f'%trainMax)
```

5000까진 괜찮은데 1만 넣으면 한나절

```python
howMaxNesti(rangeMin= 5400,rangeMax= 5600, stride= 5)
```
